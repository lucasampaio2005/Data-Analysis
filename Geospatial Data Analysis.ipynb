```python
import pandas as pd
import folium
```

```python
df_suburbs = pd.read_excel("/content/waterwatch_clean2.xlsx", sheet_name='Sheet1')

df_suburbs.head()
```

```python
map_osm = folium.Map(location=[-33.925, 18.625], zoom_start=10)
map_osm
```

```python
map_dark = folium.Map(location=[-33.925, 18.625], tiles='CartoDB dark_matter', zoom_start=10)
map_dark
```

```python
df_suburbs_locations = df_suburbs[['Latitude', 'Longitude']]

suburbs_location_list = df_suburbs_locations.values.tolist()
suburbs_location_list_size = len(suburbs_location_list)
suburbs_location_list_size
```

```python
for point in range(0, suburbs_location_list_size):
  folium.Marker(suburbs_location_list[point]).add_to(map_osm)
  folium.Marker(suburbs_location_list[point]).add_to(map_dark)
```

```python
map_osm
```

```python
for point in range(0, suburbs_location_list_size):
  folium.Marker(suburbs_location_list[point], popup=df_suburbs['Suburb'][point]).add_to(map_osm)
  folium.Marker(suburbs_location_list[point], popup=df_suburbs['Suburb'][point]).add_to(map_dark)

map_osm
```

```python
for point in range(0, suburbs_location_list_size):
  folium.Marker(suburbs_location_list[point], popup=df_suburbs['Suburb'][point], icon=folium.Icon(color='darkblue', icon_color='white', icon='tint', angle=0, prefix='fa')).add_to(map_osm)
  folium.Marker(suburbs_location_list[point], popup=df_suburbs['Suburb'][point], icon=folium.Icon(color='darkblue', icon_color='white', icon='tint', angle=0, prefix='fa')).add_to(map_dark)
```

```python
map_osm
```

```python
for point in range(0, suburbs_location_list_size):
  if df_suburbs['Oct 2017\nkl/month'][point] < 5:
    folium.Marker(suburbs_location_list[point], popup=df_suburbs['Suburb'][point], icon=folium.Icon(color='green', icon_color='white', icon='thumbs-up', angle=0, prefix='fa')).add_to(map_osm)
    folium.Marker(suburbs_location_list[point], popup=df_suburbs['Suburb'][point], icon=folium.Icon(color='green', icon_color='white', icon='thumbs-up', angle=0, prefix='fa')).add_to(map_dark)
  elif df_suburbs['Oct 2017\nkl/month'][point] > 5 and df_suburbs['Oct 2017\nkl/month'][point] <= 8:
    folium.Marker(suburbs_location_list[point], popup=df_suburbs['Suburb'][point], icon=folium.Icon(color='blue', icon_color='white', icon='thumbs-up', angle=0, prefix='fa')).add_to(map_osm)
    folium.Marker(suburbs_location_list[point], popup=df_suburbs['Suburb'][point], icon=folium.Icon(color='blue', icon_color='white', icon='thumbs-up', angle=0, prefix='fa')).add_to(map_dark)
  else:
    folium.Marker(suburbs_location_list[point], popup=df_suburbs['Suburb'][point], icon=folium.Icon(color='white', icon_color='white', icon='thumbs-down', angle=0, prefix='fa')).add_to(map_osm)
    folium.Marker(suburbs_location_list[point], popup=df_suburbs['Suburb'][point], icon=folium.Icon(color='white', icon_color='white', icon='thumbs-down', angle=0, prefix='fa')).add_to(map_dark)

map_osm
```
